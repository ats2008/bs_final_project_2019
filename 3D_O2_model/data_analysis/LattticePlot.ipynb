```python
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import *
```

```python
f=open('lattice2.txt')
l=f.readline()
lattice_spin_angles=[]
while(l):
    if "##LATTICE" in l:
        break
    l=f.readline()

l=f.readline()
while l:
    l=l[:-1].split(',')
    lattice_spin_angles.append([float(i) for i in l])
    l=f.readline()
f.close()
n,m =len(lattice_spin_angles),len(lattice_spin_angles[0])
X=[i-n/2 for i in range(n)]
Y=[i-m/2 for i in range(m)]
U,V,col=[],[],[]
for row in lattice_spin_angles:
        U.append([cos(theta) for theta in row ] )
        V.append([sin(theta) for theta in row ] )
        col.append([int(theta*256/2*3.15) for theta in row ])
print(n," x ",m)
plt.imshow(lattice_spin_angles,cmap='jet',interpolation='nearest')
plt.figure()
plt.quiver(X,Y,U,V,col,cmap=plt.get_cmap('jet'))
```

### Lattice from lattice from lattice.lat

```python
f=open('../lattice.lat','r')
l=f.readline()
lattice=[]
while l:
    if '#fstate' in l:
        lattice.append([])
        l=f.readline()
        continue
    l=l[:-1].split(',')
    lattice[-1].append([float(i) for i in l])
    l=f.readline()
f.close()
```

```python
f,ax=plt.subplots(ncols=3,nrows=2,figsize=(15,6))
i=3
i=min(i,len(lattice)-2)
i=max(0,i-1)
for idx in range(3):
    lattice_spin_angles=lattice[i]
    i+=1
    n,m =len(lattice_spin_angles),len(lattice_spin_angles[0])
    X=[i for i in range(n)]
    Y=[i for i in range(m)]
    U,V,col=[],[],[]
    for row in lattice_spin_angles:
            U.append([cos(theta) for theta in row ] )
            V.append([sin(theta) for theta in row ] )
            col.append([int(theta*256/2*3.15) for theta in row ])
    ax[0][idx].imshow(lattice_spin_angles,cmap='jet',interpolation='nearest')
    ax[1][idx].quiver(X,Y,U,V,col,cmap=plt.get_cmap('jet'),pivot='mid')
    ax[0][idx].set_title(str(i-1)+ "("+str(n)+" x "+str(m) +")")
```
